```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/nba_games_enriched.csv")
df.head()
df.info()
df.isna().sum()
# Win rate by team
win_rate = df.groupby("teamName")["win"].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=win_rate.index, y=win_rate.values)
plt.xticks(rotation=90)
plt.title("Win Rate by Team")
plt.ylabel("Win Rate")
plt.show()
plt.figure(figsize=(10,5))
sns.boxplot(x=df["win"], y=df["Q3_diff"])
plt.title("Q3 Difference vs Win")
plt.xlabel("Win (0 = Loss, 1 = Win)")
plt.ylabel("Q3 Difference")
plt.show()
plt.figure(figsize=(10,5))
sns.boxplot(x=df["win"], y=df["Foul_diff"])
plt.title("Foul Difference vs Win")
plt.xlabel("Win")
plt.ylabel("Foul Difference")
plt.show()
plt.figure(figsize=(10,5))
sns.boxplot(x=df["win"], y=df["threePA_ratio"])
plt.title("3PA Ratio vs Win")
plt.xlabel("Win")
plt.ylabel("3PA Ratio")
plt.show()
close_df = df[df["close_game"] == 1]

plt.figure(figsize=(10,5))
sns.histplot(close_df["Q3_diff"], kde=True, bins=30)
plt.title("Q3 Difference Distribution in Close Games")
plt.xlabel("Q3 Difference")
plt.show()
```
